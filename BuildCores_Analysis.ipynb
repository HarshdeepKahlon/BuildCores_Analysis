# BuildCores Data Analysis Preview
<p>This is an early preview of some general data collected in the BuildCores application for iOS and Android that will evaluate market and enthuasist trends.<p>

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_df = pd.read_csv('Data/buildcores_data.csv', error_bad_lines=False)
```

```python
data_df.info()
```

```python
for index, row in data_df.iterrows():
    newName = row["Custom parameter"].split('\r\nto')
    data_df.set_value(index, 'Custom parameter', newName[0])
```

data_df[data_df['Event count'] > 200]

```python
data_df[data_df['Event count'] > 200].plot(kind='bar',x='Custom parameter',y='Event count',color='red')
```
